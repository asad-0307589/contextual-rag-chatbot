```python
import wikipediaapi

wiki_wiki = wikipediaapi.Wikipedia(
    language='en',
    user_agent='MyChatbot/1.0 (syedmasadshah47@gmail.com)'
)

pages = [
    # Machine Learning topics
    "Artificial intelligence",
    "Machine learning",
    "Deep learning",
    "Natural language processing",
    "Neural network",
    "Support vector machine",
    "Decision tree",
    "Random forest",

    # Pakistan history & culture topics
    "History of Pakistan",
    "Pakistan Movement",
    "Independence of Pakistan",
    "Culture of Pakistan",
    "Partition of India",
    "Allama Iqbal",
    "Quaid-e-Azam Muhammad Ali Jinnah",
    "Indus Valley Civilization"
]

output_file = "combined_corpus.txt"

with open(output_file, "w", encoding="utf-8") as f:
    for page_title in pages:
        page = wiki_wiki.page(page_title)
        if page.exists():
            print(f"Fetched: {page_title}")
            f.write(f"## {page.title}\n")
            f.write(page.text + "\n\n")
        else:
            print(f"Page '{page_title}' not found on Wikipedia")
```

# Preprocessing of Dataset

```python
with open("combined_corpus.txt", "r", encoding="utf-8") as f:
    text = f.read()
```

```python
!pip install langchain
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Create splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100
)

# Make list of Document objects (each with metadata + text)
documents = text_splitter.create_documents([text])

# Or if you just want plain text chunks:
chunks = [doc.page_content for doc in documents]
```

```python
import re

clean_chunks = [re.sub(r'\s+', ' ', chunk).strip() for chunk in chunks]
```

```python
with open("Cleaned_text.txt", "w", encoding="utf-8") as f:
    for chunk in clean_chunks:
        f.write(chunk + "\n\n")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load text
with open("Cleaned_text.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()

# Split
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
documents = splitter.create_documents([raw_text])

# Use free HuggingFace embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create FAISS index
db = FAISS.from_documents(documents, embeddings)

# Save
db.save_local("faiss_index")
```

# Load the FAISS index

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

# Load embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Load the saved FAISS index
db = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
```

# HuggingFace LLM to generate answers

```python
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# Load small language model from HF
qa_pipeline = pipeline(
    "text2text-generation",
    model="google/flan-t5-small",
    tokenizer="google/flan-t5-small",
    max_length=256
)

# Wrap into LangChain LLM
llm = HuggingFacePipeline(pipeline=qa_pipeline)
```

# Connect retriever + LLM into RetrievalQA chain

```python
from langchain.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=db.as_retriever()
)
```

# Testing

```python
query = "What is artificial intelligence?"
result = qa_chain({"query": query})
print("Answer:", result["result"])
```
